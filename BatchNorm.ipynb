```python
import cv2
import numpy as np
import pylab
```

```python
#im = cv2.imread("data/047.jpg")
#file = "data/{:03d}.jpg".format(random.randint(1,404))
file = "data/047.jpg"
print(file)
im = cv2.imread(file)
im = cv2.resize(im, (0,0), fx=0.5, fy=0.5)
im = cv2.GaussianBlur(im, (5, 5), 0)
im_hsv = cv2.cvtColor(im, cv2.COLOR_BGR2HSV)
im_rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
print(im.shape)

#im_wob = im.copy()
#im_wob[:, :, 0] = 0

#display
pylab.imshow(im_rgb)
pylab.show()
print("hsv")
pylab.imshow(im_hsv[:,:,0], cmap=pylab.gray())
pylab.show()
pylab.imshow(im_hsv[:,:,1], cmap=pylab.gray())
pylab.show()
pylab.imshow(im_hsv[:,:,2], cmap=pylab.gray())
pylab.show()
```

```python
bn_im = im_rgb
bn = np.zeros_like(bn_im)
for i in range(bn_im.shape[2]):
    min_val = bn_im[:, :, i].min()
    max_val = bn_im[:, :, i].max()
    bn[:, :, i] = (bn_im[:, :, i] - min_val) * (255 / (max_val - min_val))
    pylab.imshow(np.concatenate([bn[:,:,i], bn_im[:, :, i]], axis=1), cmap=pylab.gray())
    pylab.show()
    print(bn[:, :, i].max(), bn[:, :, i].min())
pylab.imshow(np.concatenate([bn, bn_im], axis=1), cmap=pylab.gray())
pylab.show()
```

```python
hsv_bn = cv2.cvtColor(bn, cv2.COLOR_RGB2HSV)
for i in range(hsv_bn.shape[2]):
    pylab.imshow(np.concatenate([hsv_bn[:,:,i], im_hsv[:, :, i]], axis=1), cmap=pylab.gray())
    pylab.show()
```
